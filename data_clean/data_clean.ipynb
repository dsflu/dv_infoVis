```python
import pandas as pd
df2 = pd.read_excel('gtd_12to15_0616dist.xlsx', encoding='utf8')
df1 = pd.read_excel('gtd_92to11_0616dist.xlsx', encoding='utf8')
df =pd.concat([df1,df2])


# def filter_level(nkill,nwound)
#     if nkill > 100:
#         return True
#     else:
#         return False

```

```python
df.head()
```

```python
ratio_kill = 0.6
ratio_wound = 1.0 - ratio_kill

df["level"] = df.nkill*ratio_kill + df.nwound*ratio_wound
mask = df.level > 20

```

```python
df['time'] = df.apply(lambda x : str(x['iyear'])+'-'+str(x['imonth'])+'-'+str(x['iday'])+' 00:00:00',axis=1)
df['year_m'] = df.apply(lambda x : x['iyear']+x['imonth']/12,axis=1)
df.time.head()
df_new = df[mask]
```

```python
filter_data = ["eventid",'time','year_m',"iyear","imonth","iday","region","country_txt","city","latitude","longitude","attacktype1","targtype1","nkill","nwound","level"]
df_filter = df_new[filter_data]
df_filter.tail(20)
```
